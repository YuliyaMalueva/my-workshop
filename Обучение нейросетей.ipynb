```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(17)
tf.random.set_seed(17)
```

```python
df = pd.read_csv('prepared_posts_df.csv')
df.head()
```

```python
X = df.drop(columns=['rating']).to_numpy()
Y = df['rating'].to_numpy()
```

```python
val_frac = 0.15
test_frac = 0.15

x_train = X[:round((1 - val_frac - test_frac) * len(X))]
x_val = X[round((1 - val_frac - test_frac) * len(X)):round((1 - test_frac) * len(X))]
x_test = X[round((1 - test_frac) * len(X)):]

y_train = Y[:round((1 - val_frac - test_frac) * len(Y))]
y_val = Y[round((1 - val_frac - test_frac) * len(Y)):round((1 - test_frac) * len(Y))]
y_test = Y[round((1 - test_frac) * len(Y)):]

x_train.shape, x_val.shape, x_test.shape
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(100, activation='relu', input_shape=(X.shape[1],)),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(1, activation=tf.keras.activations.linear)
])
```

```python
# Загрузка весов
model.load_weights('homework_init_weights.h5')
```

```python
# Компиляция модели с метрикой MAE
model.compile(optimizer='adam', 
              loss='mse', 
              metrics=['mae'])

# Вычисление MAE на всем датасете
loss, mae = model.evaluate(X, Y, verbose=0)
```

```python
print(f"Mean Absolute Error (MAE): {mae}")
print(f"MAE округленный до целого: {int(round(mae))}")
```

```python
model.compile(optimizer='adam', 
              loss='mse', 
              metrics=['mae'])
```

```python
history = model.fit(X, Y,
                   batch_size=5000,
                   epochs=300,
                   verbose=1)
```

```python
final_loss = history.history['loss'][-1]
```
