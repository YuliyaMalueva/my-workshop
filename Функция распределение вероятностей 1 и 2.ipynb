```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sns
```

```python
data = np.random.randint(10, 101, size=20)
data
```

```python
plt.hist(data, bins=20, edgecolor='black')
plt.title('Гистограмму плотности распределения')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.show()
```

```python
plt.hist(data, histtype='step', cumulative=True, bins=len(data))
plt.title('График эмпирической плотность распределения')
```

```python
bin_dt, bin_gr = np.histogram(data, bins=len(data))
Y = bin_dt.cumsum()
for i in range(len(Y)):
    plt.plot([bin_gr[i], bin_gr[i+1]],[Y[i], Y[i]],color='green')
```

```python
sns.kdeplot(data, fill=True)

plt.title('Эмпирическая плотность распределения')
plt.xlabel('Значения')
plt.ylabel('Плотность')
plt.grid()

plt.show()
```

```python
# Вычисляем точечную оценку математического ожидания
mean_estimate = np.mean(data)

print("Точечная оценка математического ожидания:", mean_estimate)
```

```python
# Расчет точечной оценки дисперсии
variance = np.var(data, ddof=0) 
print("Точечная оценка дисперсии:", variance)
```

```python
def f(x):
    return (2/3) * (1 - x)

x = np.linspace(0, 1, 100)
y = f(x)
```

```python
#Нарисуйте график плотности распределения
plt.plot(x, y)
plt.fill_between(x, y, alpha=0.5)
plt.title('График плотности распределения')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

```python
#Нарисуйте график интегральной функции распределения, f(x)=0

def F(x):
    return (2/3) * (x - (x*2)/2)  

x = np.linspace(0, 1, 100)
y = F(x)

plt.plot(x, y, label='CDF F(x)')
plt.title('График интегральной функции распределения')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.show()
```
