```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
plt.hist(df['SepalLengthCm'], bins=20, edgecolor='black')
plt.title('Гистограмма SepalLengthCm')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.show()
```

```python
# Вычисляем точечную оценку математического ожидания
mean_estimate = np.mean(df['SepalLengthCm'])
print("Точечная оценка математического ожидания:", mean_estimate)

# Расчет точечной оценки дисперсии
variance = np.var(df['SepalLengthCm'], ddof=0) 
print("Точечная оценка дисперсии:", variance)
```
