```python
# в NLTK уже имеются списки стоп-слов для разных языков
# давайте им воспользуемся
from nltk.corpus import stopwords

# токенизацию будем выполнять при помощи функции word_tokenize
from nltk.tokenize import word_tokenize

# будем осуществлять стеммизацию слов при помощи RussianStemmer
from nltk.stem.snowball import RussianStemmer
```

```python
stopwords.words('russian')
```

```python
stop_words = set(stopwords.words('russian'))
```

```python
stemmer = RussianStemmer()
```

```python
def preprocess(comment):
    # сначала комментарий преобразуем к нижнему регистру
    # заглавные буквы нам ни к чему
    comment = comment.lower()

    # токенизируем текст, получив список токенов
    tokens = word_tokenize(comment)

    # в списке оставим только те токены, что являются словами (выкинув, например, знаки препинания)
    # также выкинем стоп-слова
    # всё, что осталось, подвергнем стеммизации
    tokens = [stemmer.stem(t) for t in tokens if t.isalpha() and t not in stop_words]

    return tokens
```

```python
comments = [preprocess(comm) for comm in comments]
```

```python
import gensim
```

```python
model = gensim.models.Word2Vec(
    sentences=comments,
    vector_size=50,  # вместо size
    window=5,
    min_count=5,
    sg=1,
    epochs=50,       # вместо iter
    workers=4
)
```

```python
model.save('comments_model.w2v')
```

```python
def get_vocab_info(model):
    """Возвращает полную информацию о словаре в старом формате"""
    vocab_info = {}
    
    for word in model.wv.key_to_index.keys():
        # Создаем объект со всеми атрибутами старого Vocab
        class Vocab:
            def __init__(self, index, count):
                self.index = index
                self.count = count
                self.sample_int = 0  # для совместимости
            
            def __repr__(self):
                return f"<gensim.models.keyedvectors.Vocab at {hex(id(self))}>"
        
        vocab_info[word] = Vocab(
            index=model.wv.key_to_index[word],
            count=model.wv.get_vecattr(word, 'count')
        )
    
    return vocab_info

# Использование
vocab = get_vocab_info(model)
print("Словарь создан в старом формате")
print(f"Пример: 'поч' -> {vocab['поч']}")
```
