```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Создаем модель той же архитектуры
model = Sequential()
model.add(Dense(60, input_dim=59, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='linear'))

# Загружаем веса из файла
model.load_weights('weights.h5')

# Создаем входной вектор из одних единиц
input_vector = np.repeat(1, 59)[np.newaxis]

# Получаем предсказание
prediction = model.predict(input_vector)

# Округляем до целого
result = int(round(prediction[0][0]))
print(result)
```
