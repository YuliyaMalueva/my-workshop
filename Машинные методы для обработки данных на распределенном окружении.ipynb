```python
from sklearn.datasets import load_wine

# Загрузка wine dataset
data = load_wine()
X, y = data.data, data.target

print(f"Данные (X): {X.shape}")
print(f"Метки (y): {y.shape}")
print(f"Названия признаков: {data.feature_names}")
print(f"Названия классов: {data.target_names}")
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression

dataset = spark.read.csv('WineRecognition.csv',inferSchema=True, header=True)
```

```python
df.dtypes
```

```python
dataset.printSchema()
```

```python
# Input all the features in one vector column
assembler = VectorAssembler(
    inputCols=['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash',
               'magnesium', 'total_phenols', 'flavanoids', 'nonflavanoid_phenols',
               'proanthocyanins', 'color_intensity', 'hue', 'od280_od315', 'proline'],
    outputCol='Features')

output = assembler.transform(dataset)

# Input vs Output
finalized_data = output.select("Features", "class")

finalized_data.show()
```

```python
# Split training and testing data
train_data, test_data = finalized_data.randomSplit([0.8, 0.2], seed=42)

# Используем классификатор
classifier = LogisticRegression(featuresCol='Features', labelCol='class')

# Learn to fit the model from training set
model = classifier.fit(train_data)

# To predict on testing set
predictions = model.transform(test_data)

# Predict the model
predictions.select("Features", "class", "prediction", "probability").show(truncate=False)
```

```python
# Для многоклассовой логистической регрессии используем coefficientMatrix и interceptVector
coefficient_matrix = model.coefficientMatrix
intercept_vector = model.interceptVector

print("The coefficient matrix of the model is:")
print(coefficient_matrix)
print(f"\nThe intercept vector of the model is: {intercept_vector}")
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Создаем evaluator для многоклассовой классификации
evaluator = MulticlassClassificationEvaluator(labelCol="class", predictionCol="prediction")

# Accuracy
accuracy = evaluator.evaluate(predictions)
print("Accuracy: %.3f" % accuracy)

# F1 Score
f1 = evaluator.evaluate(predictions, {evaluator.metricName: "f1"})
print("F1 Score: %.3f" % f1)

# Weighted Precision
weighted_precision = evaluator.evaluate(predictions, {evaluator.metricName: "weightedPrecision"})
print("Weighted Precision: %.3f" % weighted_precision)

# Weighted Recall
weighted_recall = evaluator.evaluate(predictions, {evaluator.metricName: "weightedRecall"})
print("Weighted Recall: %.3f" % weighted_recall)
```
