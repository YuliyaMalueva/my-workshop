```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
lam = 14

mean = lam  # математическое ожидание
variance = lam  # дисперсия для распределения Пуассона
std_dev = np.sqrt(lam)  # стандартное отклонение

print(f'Математическое ожидание (mean): {mean}')
print(f'Дисперсия (variance): {variance}')
print(f'Стандартное отклонение (std_dev): {std_dev}')
```

```python
# Гистограмма относительных частот
x = np.arange(0, 40) 
pmf = poisson.pmf(x, lam)

plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.bar(x, pmf, alpha=0.7, color='blue', label='PMF')
plt.title('Гистограмма относительных частот распределения Пуассона (λ=14)')
plt.xlabel('Число событий')
plt.ylabel('Относительная частота')
plt.xticks(x)
plt.legend()
```

```python
# График кумулятивной функции распределения
cdf = poisson.cdf(x, lam)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, marker='o', linestyle='-', color='blue', label='CDF')
plt.title('Кумулятивная функция распределения Пуассона (λ=14)')
plt.xlabel('Число событий')
plt.ylabel('Кумулятивная вероятность')
plt.xticks(x)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np

a = 0
b = 5

# Среднее время ожидания
mean_waiting_time = (a + b) / 2

# Среднеквадратическое отклонение времени ожидания
std_deviation = np.sqrt((b - a) ** 2 / 12)

mean_waiting_time, std_deviation
print(f'Среднее время ожидания: {mean_waiting_time}')
print(f'Среднеквадратическое отклонение времени ожидания: {std_deviation}')
```

Задание 2.2

```python
import scipy.stats as stats
```

```python
# 1. Закон распределения случайной величины Х – количество обслуженных заявителей в час
lambda_per_hour = 3
lambda_per_2_hours = lambda_per_hour * 2
k_values = np.arange(0, 15)

# Вычисление вероятностей
probabilities = poisson.pmf(k_values, lambda_per_hour)

# Вывод результатов
for k, p in zip(k_values, probabilities):
    print(f'P(X = {k}) = {p:.4f}')
```

```python
# 2. Вероятность того, что за 2 часа будет обслужено 8 заявителей
prob_8_in_2_hours = stats.poisson.pmf(8, lambda_per_2_hours)
print(f"Вероятность того, что за 2 часа будет обслужено 8 заявителей: {prob_8_in_2_hours:.4f}")
```

```python
# 3. Вероятность того, что за 2 часа будет обслужено более 10 заявителей
prob_more_than_10_in_2_hours = 1 - stats.poisson.cdf(10, lambda_per_2_hours)
print(f"Вероятность того, что за 2 часа будет обслужено более 10 заявителей: {prob_more_than_10_in_2_hours:.4f}")
```

Задание 2.3

```python
# Параметр λ
lambda_param = 5 * 0.05  # Ожидаемое число опозданий
print(f"λ (ожидаемое количество опозданий): {lambda_param}")
```

```python
# Вероятность того, что ни один поезд не опоздает
P_0 = stats.poisson.pmf(0, lambda_param)
print(f"Вероятность того, что ни один поезд не опоздает: {P_0:.4f}")
```

```python
# Вероятность того, что опоздают ровно 2 поезда
P_2 = stats.poisson.pmf(2, lambda_param)
print(f"Вероятность того, что опоздают ровно 2 поезда: {P_2:.4f}")
```

```python
# Вероятность того, что опоздают хотя бы 2 поезда
P_at_least_2 = 1 - stats.poisson.cdf(2, lambda_param)
print(f"Вероятность того, что опоздают хотя бы 2 поезда: {P_at_least_2:.4f}")
```
