```python
df = pd.DataFrame({
    'id':[],
    'book_name':[],
    'price':[],
    'url':[],
})
```

```python
def parse_book_details(book_url):
    response = requests.get(book_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    upc = soup.find('table', class_='table').find_all('tr')[0].find_all('td')[0].text
    return upc
```

```python
def get_product_pod(soup, data=df):
    
    articles = soup.find_all('article', attrs={'class':'product_pod'})
    
    for article in articles:
        product_pod = {}
        product_pod['book_name'] = article.find('img').get('alt')
        product_pod['price'] = article.find('p', class_='price_color').text
        product_pod['url'] = 'https://books.toscrape.com' + article.find('a').get('href')

        product_pod['id'] = parse_book_details()
        df.loc[len(df)] = product_pod
```

```python
url = 'https://books.toscrape.com/page/'
i = 1

while True:
    print(f'Parsing page N:{i}')
    r = requests.get(url + str(i))
    soup = BeautifulSoup(r.content, 'html.parser')
    
    if 'No product_pod found!' in soup.text:
        break
    i += 1

print('Parsing done!')
```
